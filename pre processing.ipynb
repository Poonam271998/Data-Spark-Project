```python
import pandas as pd


not_null_df['Order Date'] = pd.to_datetime(not_null_df['Order Date'])
not_null_df['Delivery Date'] = pd.to_datetime(not_null_df['Delivery Date'])

# Calculate the duration in days for rows with non-null Delivery Date
not_null_df['Duration'] = (not_null_df['Delivery Date'] - not_null_df['Order Date']).dt.days

# Calculate the average duration for non-null Delivery Dates
average_duration = not_null_df['Duration'].mean()

print("Average Duration overall:", average_duration)
average_duration_by_currency = not_null_df.groupby('Currency Code')['Duration'].mean()
print("Average duration group by currency\n" ,average_duration_by_currency)

mode_duration = not_null_df['Duration'].mode()
print("mode Duration overall:", mode_duration[0])
```
